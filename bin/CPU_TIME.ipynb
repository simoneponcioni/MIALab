```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# font stix
plt.rcParams["mathtext.fontset"] = "stix"
plt.rcParams["font.family"] = "STIXGeneral"
plt.rcParams["font.size"] = 16

# List1 
lst = [['DICE', 0.00226], ['HD95', 0.07191], ['SBD', 0.13646]]
    
df = pd.DataFrame(lst, columns =['Metric', 'CPU Time (s)'])

plt.figure(figsize=(8,8))
sns.barplot(x="Metric", y="CPU Time (s)", data=df, palette="Set3", linewidth=1)
plt.title("CPU Time (s)", weight="bold", fontsize=20)
# add value on top of bar
for index, row in df.iterrows():
    plt.text(row.name,row['CPU Time (s)'], round(row['CPU Time (s)'], 5), color='black', ha="center")
plt.tight_layout()
plt.savefig("/home/simoneponcioni/Documents/03_LECTURES/MIALab/sbd_results/cpu_time.png")
plt.show()
```
